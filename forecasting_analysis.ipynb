```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
results_auto = []
results_auto_14 = []
results_arima = []
results_dummy = []
results_dummy_mean = []
for test_idx in range(8,70):
    results_auto.append(pd.read_pickle('results/autoregresssive_{}.pkl'.format(test_idx)))
    results_auto_14.append(pd.read_pickle('results/autoregresssive-14_{}.pkl'.format(test_idx)))
    results_arima.append(pd.read_pickle('results/arima_{}.pkl'.format(test_idx)))
    results_dummy.append(pd.read_pickle('results/dummy_prev_week_{}.pkl'.format(test_idx)))
    results_dummy_mean.append(pd.read_pickle('results/dummy_mean_week_{}.pkl'.format(test_idx)))
```

```python
def summarize_for_plotting(res):
    indicator_1 = []
    indicator_2 = []
    indicator_3 = []
    for i, df in enumerate(res):
        indicator_1.append(df['PI1'])
        indicator_2.append(df['PI2'])
        indicator_3.append(df['PI3'])
    return (np.vstack(indicator_1), np.vstack(indicator_2), np.vstack(indicator_3))
```

```python
results_auto = summarize_for_plotting(results_auto)
results_auto_14 = summarize_for_plotting(results_auto_14)
results_arima = summarize_for_plotting(results_arima)
results_dummy = summarize_for_plotting(results_dummy)
results_dummy_mean = summarize_for_plotting(results_dummy_mean)
```

```python
metrics_id = 1
for i in range(10):
    plt.figure()
    plt.plot(results_auto[metrics_id][:,i], label='auto')
    plt.plot(results_auto_14[metrics_id][:,i], label='auto-14')
    plt.plot(results_arima[metrics_id][:,i], label='arima')
    plt.plot(results_dummy[metrics_id][:,i], label='dummy')
    plt.plot(results_dummy_mean[metrics_id][:,i], label='mean')
    plt.title('DMA {}'.format(i+1))
    plt.legend()
    plt.ylim((0,15))
    plt.show()
```
