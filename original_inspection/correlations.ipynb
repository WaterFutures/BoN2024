```python
# Import the necessary external libraries
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
```

```python
#Â simple correlation betwen signals 
import statsmodels.tsa.stattools as stools

from preprocessing.detrend_mnf import Detrend
dmas_h_q = Detrend().fit_transform(dmas_h_q)

fig = make_subplots(rows=2, cols=1, shared_xaxes=True)

for i, dma in enumerate(dmas_h_q.columns):
    acf_res, acf_confint = stools.acf(dmas_h_q[dma], 
                                         nlags=WEEK_LEN+2,
                                         alpha=0.05,
                                         missing='drop')

    fig.add_trace(go.Bar(x=np.arange(len(acf_res)), 
                             y=acf_res, 
                             name=dma, 
                             legendgroup=dma,
                             showlegend=True,
                             base=dict(color=colors[i]))
                             , row=1, col=1)
    
for i, dma in enumerate(dmas_h_q.columns):
    pacf_res, pacf_confint = stools.pacf(dmas_h_q[dma].interpolate(), 
                                         nlags=WEEK_LEN+2,
                                         alpha=0.05,
                                         missing='drop')

    fig.add_trace(go.Bar(x=np.arange(len(pacf_res)), 
                             y=pacf_res, 
                                name=dma,
                                showlegend=False,
                             legendgroup=dma,
                             base=dict(color=colors[i])),
                             row=2,col=1)
    

fig.show()

# lagged correlation between signals ? 
```
